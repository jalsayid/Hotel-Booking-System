```python
# This work done by group 2:
# Joud Alsayid, 50%
# Sara Albajad, 50%


from random import *
## This function will keep running untill the user exit it with option 6 
# It provides the user with a number of choices to ease the process of booking
#
def main() :
    flag = True
    # As soon as the flag = True this loop will run
    while flag :
        # To display the main menu
        menu()
        print()
        # The user will choose an option
        choice = input("Enter your choice(a number): ")
        # To handle exceptions with the try/except statement 
        try :
# The if statement allows a program to carry out different actions depending on the nature of the data to be processed
            if choice == "1" :
            # To collect the dict containing the users data
                data = newHotelBooking()
            elif choice == "2" :
                bookingConfirmation(data)
            elif choice == "3" :
                historyOfBookings()
            elif choice == "4" :
                userData = input("Enter Confirmation number or phone number: ")
                specificBooking(userData)
            elif choice == "5" :
                updatingBookingsMenu()
            elif choice == "6" :
                # To stop the loop
                flag = False
            else :
            # To signal an exceptional condition, use the raise statement
                raise TypeError
        except TypeError :
            print("Error: Invalid input, try again")


## Display the menu of the hotel booking system
#
def menu() :
    print("#" * 70)
    print("%45s" % "Hotel Booking System")
    print("%32s" % "1. Add New Hotel Booking")
    print("%54s" % "2. Generate Booking Confirmation (as txt file)")
    print("%47s" % "3. Display Information for all Bookings")
    print("%36s" % "4. Display Specific Booking ")
    print("%37s" % "5. Update Booking Information")
    print("%15s" % "6. Exit")
## This function allows the user to add a new Booking in the system.
# The information is saved in the Bookings.txt file.
# @return bookingData(dictionary) enterd from the user this data is saved in a dictionary only if it is valid
#
def newHotelBooking() :
    try :
        # To obtain data from the user
        fullName = input("Enter your Full name : ")
        # Checks if a name is valid
        if len(fullName.split(" ")) == 1:
            raise ValueError
        # Checks if a number is valid : its need to contain 10 numbers and to start with 05
        phoneNumber = input("Enter your phone number : ")
        if len(phoneNumber) != 10 or phoneNumber[:2] != "05" :
            raise ValueError
        city = input("Enter your city name : ")
        hotel = input("Enter your hotel name : ")
        checkIn =  input("Enter your check-in date(day/month/year) : ")
        checkOut =  input("Enter your check-out date(day/month/year) : ")
        # To store the date in tuples
        day1, month1, year1 = checkIn.split("/")
        day2, month2, year2 = checkOut.split("/")
        # Checks if a check-in/check-out date is valid
        # the days need to be between 30 and 1
        if  not(0 < int(day1) <= 30) and not(0 < int(day2) <= 30) :
            raise ValueError
        # the month should be between 12 and 1
        if  not(0 < int(month1) <= 12) and not(0 < int(month2) <= 12) :
            raise ValueError
        # the year should be greater than 2022
        if  int(year1) < 2022 and int(year2) < 2022 :
            raise ValueError
        # the check-out date should be after the check-in date
        if int(year2) < int(year1) :
            raise ValueError
        elif year1 == year2 :
            if int(month2) < int(month1) :
                raise ValueError
            elif month1 == month2 :
                if int(day2) <= int(day1) :
                    raise ValueError
        numberOfPeople = input("Enter the number of people : ")
        # initialise the varible digits
        digits = ""
        # loop to generate a random number for the confirmation number
        for i in range(4) :
            num = str(randint(0,9))
            digits += num
        confirmationNumber = city[0].upper() + fullName[0].upper() + digits
        # Storing the users data in the dict--> bookingData
        bookingData = {
            "Full Name" : fullName, 
            "Phone Number" : phoneNumber,
            "City Name" : city,
            "Hotel Name" : hotel,
            "Check-in Date" : checkIn, 
            "Check-out Date" : checkOut,
            "Number of People" : numberOfPeople, 
            "Confirmation Number" : confirmationNumber
        }
        # adding data in the bookings.txt file 
        bookingInfo = fullName+" "+phoneNumber+" "+city+" "+hotel+" "+checkIn+" "+checkOut+" "+numberOfPeople+" "+confirmationNumber
        dataFile = open("Bookings.txt", "a")
        print(bookingInfo, file = dataFile)
        # To close the file
        dataFile.close()
        
        return bookingData
    
    except ValueError :
        print("Invalid data")
## Generates the confirmation letter for the booking as a file named “confirmation_lastname.txt”
# @param data is the dictionary containing the last booking information from the user
# 
def bookingConfirmation(data):
    # To extract the last name for the file name
    Name = data["Full Name"].split()
    lastName = Name[1]
    fileName = "confirmation" + "_" + lastName +".txt"
    bookingFile = open(fileName, "w")
    bookingFile.write("%45s" % "Your booking is confirmed")
    star = "*" * 120
    bookingFile.write("\n%s" % star)
    bookingFile.write("\n")
    # to display the information in the file
    for key in data.keys() :
        print(key,":",data[key], file = bookingFile)
    bookingFile.write("%s" % star)
    bookingFile.close()

## This function allows the user to display all the Hotel Booking’s information in a nicely formatted table
#
def historyOfBookings() :
    print('-'*91)
    print("%-8s%7s%11s%11s%9s%11s%13s%13s" % ('Confirmation', 'Name', 'Phone', 'City', 'Hotel', 'Check-in','Check-out', 'Number'))
    print("%-8s%49s%12s%21s" % ("number","date", "date", "of Guests" ))
    print("-"*91)
    dataFile = open("Bookings.txt", "r") # opened the file in read mode to extract the data
    for line in dataFile :
        line = line.split()
        print("%-11s%-14s%-12s%-9s%-8s%-12s%-17s%-s" % (line[8], line[0]+" "+line[1], line[2], line[3], line[4],line[5],line[6],line[7]))
## This function allows the user to display a specific booking information based on Confirmation Number or Phone number.
# @param userInput Confirmation Number or Phone number that was entered by the user 
#
def specificBooking(userInput) :
    # If the Confirmation Number/phone number is not found an error message will appear
    notFound = True
    dataFile = open("Bookings.txt", "r")
    while notFound :
        # we put the data in the list line to ease the proccess of collecting data
        line = dataFile.readline().split()
        if userInput in line :
            notFound = False
        if line == [] :
            notFound = False
    if line != [] :
        print("Confirmation Number: ", line[8])
        print("Phone Number: ",  line[2])
        print("City: ", line[3])
        print("Hotel: ", line[4])
        print("Check-in: ", line[5])
        print("Check-out: ", line[6])
        print("Number of Guests: ", line[7])
    else :
        print("Entered Confirmation Number or Phone number is wrong")
        
    dataFile.close()
# This function display a menu to update Booking or cancel it
def updatingBookingsMenu() :
    print("%37s" % "1. Update Booking information")
    print("%43s" % "2. Cancel Hotel Booking information")
    print("%32s" % "3. Back to the main menu")
    userChoice = input("Enter your choice: ")
    if userChoice == "1" :
        updateBookings()
    elif userChoice == "2" :
        cancelBooking()

    # else if the user entered "3" he will get back to the main menu using the main function
    
# This function should allow the user to change the hotel data using confirmation number
def updateBookings() :
    # To read the information added to bookings.txt
    dataFile = open("Bookings.txt", "r")
    # To read the hole file and put each line in a list
    data = dataFile.readlines()
    # The user will have the choice to update the needed data using the confirmation number
    confNum = input("Enter your confirmation number to update your booking: ")
    print("%28s" % "1. update hotel name")
    print("%45s" % "2. update check-in and check-out date")
    print("%22s" % "3. update both")
    choice = input("Enter your choice: ")
    # i is each line in the file stored in the list data
    for i in data :
        # if the confirmation number was found in the line this line will be updated
        if confNum in i :
            line = i.rstrip()
            # removing the line to update it with a new line containing updated data
            data.remove(i)
    # To covnvert the line to a list to modify it
    line = line.split()
    dataFile.close()
    # writing the new data to the file
    dataFile = open("Bookings.txt", "w")
    if choice == "1" :
        newHotel = input("Enter your new hotel name: ")
        # adding new data
        line[4] = newHotel
        for i in data :
            print(i.rstrip(), file = dataFile)
        # i is each information available in the line
        for i in line :
            print(i, end = " ",  file = dataFile)
            
    elif choice == "2" :
        checkIn =  input("Enter your new check-in date(day/month/year) : ")
        checkOut =  input("Enter your new check-out date(day/month/year) : ")
        # adding new data
        line[5] = checkIn
        line[6] = checkOut
        for i in data :
            print(i.rstrip(), file = dataFile)
        # i is each information available in the line
        for i in line :
            print(i, end = " ",  file = dataFile)
        
    else :
        newHotel = input("Enter your new hotel name: ")
        checkIn =  input("Enter your new check-in date(day/month/year) : ")
        checkOut =  input("Enter your new check-out date(day/month/year) : ")
        # adding new data
        line[4] = newHotel
        line[5] = checkIn
        line[6] = checkOut
        for i in data :
            print(i.rstrip(), file = dataFile)
        # i is each information available in the line
        for i in line :
            print(i, end = " ",  file = dataFile)
    
    dataFile.close()

## This function allows the user to delete booking information using confirmation number 
#
def cancelBooking() :
    confNum = input("Enter your confirmation number to cancel your booking: ")
    dataFile = open("Bookings.txt", "r")
    # To read the hole file and put each line in a list
    data = dataFile.readlines()
    # i is each line in the file stored in the list data
    for i in data :
        if confNum in i :
            data.remove(i)
    dataFile.close()
    # writing the data with cancellation to the file
    dataFile = open("Bookings.txt", "w")
    for i in data :
        print(i.rstrip(), file = dataFile)
    dataFile.close()
# Calling the main function to start the program/project    
main()   
```
